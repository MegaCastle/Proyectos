```python
x= df.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans= KMeans(n_clusters= i, init = "k-means++", max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("Métofdo del codo")
plt.xlabel("Número de Clusters")
plt.ylabel("WCSS(k)")
plt.show()
```

```python
#Segmentación del data set por k-means
kmeans= KMeans(n_clusters=5, init="k-means++", max_iter=300, n_init=10, random_state=0)
y_kmeans=kmeans.fit_predict(x)
y_kmeans
```

```python
#Visualización
plt.scatter(x[y_kmeans==0,0],x[y_means==0,1],s=100, c="red", label="Estandar 1")
plt.scatter(x[y_kmeans==1,0],x[y_means==1,1],s=100, c="blue", label="Objetivo")
plt.scatter(x[y_kmeans==2,0],x[y_means==2,1],s=100, c="green", label="Cautos")
plt.scatter(x[y_kmeans==3,0],x[y_means==3,1],s=100, c="cyan", label="Conservadores")
plt.scatter(x[y_kmeans==4,0],x[y_means==4,1],s=100, c="magenta", label="Descuidados")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300, c="yellow", label="Baricentros")
plt.title("Cluster de clientes")
plt.xlabel("Ingresos anuales(miles de $)")
plt.ylabel("Puntuación de Gastos (1-100)")
plt.legend()
plt.show()
```
